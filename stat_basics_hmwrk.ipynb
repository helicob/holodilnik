```python
# смотрю проценты пропусков по колонкам
for col in df1.columns:
    pct_missing = df[col].isnull().mean()
    print(f'{col} - {pct_missing :.1%}')
```

```python
# 1: surgery -- номинальные данные
df1['1: surgery'] = df1['1: surgery'].apply(lambda x: True if x == 1 else x == False)
df1
```

```python
df1.describe()
# 2: Age -- количественные, дискретные, выбросов нет
# 3: Hospital Number -- номинальные данные, на всякий случай в строку перевожу
df1['3: Hospital Number'] = df1['3: Hospital Number'].apply(str)
```

```python
df1.describe()
# 4: rectal temperature -- количественные, непрерывные. пропуски забью медианой, выбросов вроде нет
df1['4: rectal temperature'] = df1['4: rectal temperature'].fillna(value=df['4: rectal temperature'].median())
# 5: 11: pain -- номинальные данные, пропуски буду считать отсутствием боли и на всякий случай 
#                в строку перевожу через int
df1['11: pain'].replace(np.nan, 1, inplace=True)
df1['11: pain'] = df1['11: pain'].apply(int)
df1['11: pain'] = df1['11: pain'].apply(str)
```

```python
df1.describe()
# 13: abdominal distension -- качественные, порядковые. заполнять пропуски не хочу. ну или разве что модой
print(df1['13: abdominal distension'].mode())
print(df1['13: abdominal distension'].mean())
print(df1['13: abdominal distension'].median())
```

```python
# 19: packed cell volume -- количественные, непрерывные. вижу, что выброс есть, но он как мне кажется,
#                           вполне может быть не ошибкой данных. пропуски наверное заполню медианой.
q1 = df1['19: packed cell volume'].quantile(0.25)
q3 = df1['19: packed cell volume'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)

print(lower_bound)
print(upper_bound)
print(df1['19: packed cell volume'].mode())
print(df1['19: packed cell volume'].mean())
print(df1['19: packed cell volume'].median())
df1['19: packed cell volume'] = df1['19: packed cell volume'].fillna(value = df1['19: packed cell volume'].median())
```

```python
# 20: total protein -- количественные, непрерывные. выбросов вроде нет.
#                      пропуски наверное заполню медианой.
q1 = df1['20: total protein'].quantile(0.25)
q3 = df1['20: total protein'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)

print(lower_bound)
print(upper_bound)
print(df1['20: total protein'].mode())
print(df1['20: total protein'].mean())
print(df1['20: total protein'].median())
df1['20: total protein'] = df1['20: total protein'].fillna(value = df1['20: total protein'].median())
```
