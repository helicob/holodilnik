## Задание 1

Напишите функцию, которая принимает на вход строку и проверяет является ли она валидным транспортным номером (1 буква, 3 цифры, 2 буквы, 2-3 цифры). Обратите внимание, что не все буквы кириллического алфавита используются в транспортных номерах.

Если номер валиден, то функция должна возвращать отдельно номер и регион.

Примеры работы программы:  

1. `car_id = 'A222BC96'`

Результат: 
`Номер A222BС валиден. Регион: 96`

2. `car_id = 'АБ22ВВ193'`

Результат: 
`Номер не валиден`

````python
import re

car_id = 'A222BC96'

def chek_num_is_valid(string):
    strcheck = re.match(r'^[АаВвЕеКкМмНнОоРрСсТтУуХхAaBbEeKkMmHhOoPpCcTtYyXx]\d{3}[АаВвЕеКкМмНнОоРрСсТтУуХхAaBbEeKkMmHhOoPpCcTtYyXx]{2}\d{2}$', string)
    if strcheck != None:
        return f'Номер {strcheck.group()[:6]} валиден. Регион: {strcheck.group()[-2:]}'
    else:
        return 'Номер не валиден'

print(chek_num_is_valid(car_id))
````

## Задание 3

Напишите функцию, которая будет возвращать акроним по переданной в нее строке со словами.

Примеры работы программы:  

1. `some_words = 'Информационные технологии'`

Результат: 
`ИТ`

2. `some_words = 'Near Field Communication'`

Результат: 
`NFC`

````python
some_words = 'Информационные технологии'
some_words1 = 'Near Field Communication'

def make_acronym(string):
    acr = re.findall(r'\b\w', string)
    return ''.join(acr).upper()

print(make_acronym(some_words1))
````

## Задание 4

Напишите функцию, которая будет принимать на вход список email-адресов и выводить их распределение по доменным зонам.

Пример работы программы:  

`emails = ['test@gmail.com', 'xyz@test.in', 'test@ya.ru', 'xyz@mail.ru', 'xyz@ya.ru', 'xyz@gmail.com']`

Результат: 
```
gmail.com: 2
test.in: 1 
ya.ru: 2 
mail.ru: 1
```

````python
emails = ['test@gmail.com', 'xyz@test.in', 'test@ya.ru', 'xyz@mail.ru', 'xyz@ya.ru', 'xyz@gmail.com']

def domain_counter(emailslist):
    res = {}

    for email in emailslist:
        dom = re.search(r'@\w+\.\w+', email).group(0)[1:]
        if dom not in res:
            res[dom] = 1
        else:
            res[dom] += 1

    for key, value in res.items():
        print(f'{key}: {res[key]}')
        
domain_counter(emails)
````

Пример работы программы:  

`some_text = 'Эталонной реализацией Python является интерпретатор CPython, поддерживающий большинство активно используемых платформ. Он распространяется под свободной лицензией Python Software Foundation License, позволяющей использовать его без ограничений в любых приложениях, включая проприетарные.'`

Результат: 
```
Слов на гласные буквы: 9
Слов на согласные буквы: 21 
```

````python
some_text = 'Эталонной реализацией Python является интерпретатор CPython, поддерживающий большинство активно используемых платформ. Он распространяется под свободной лицензией Python Software Foundation License, позволяющей использовать его без ограничений в любых приложениях, включая проприетарные.'

def start_from_vowel_not_vowel_counter(string):
    vowelwordscounter = len(re.findall(r'\b[АаУуОоИиЭэЫыЯяЮюЕеЁёAaEeIiOoUuYy]', string)) # с гласных
    notvowelwordscounter = len(re.findall(r'\b\w+', string)) - vowelwordscounter # все слова минус те, что с гласных
    print(f'Слов на гласные буквы: {vowelwordscounter}')
    print(f'Слов на согласные буквы: {notvowelwordscounter}')

start_from_vowel_not_vowel_counter(some_text)
````
