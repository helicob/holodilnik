# Домашнее задание к лекции "Функции"

Вам нужно помочь секретарю автоматизировать работу. Для этого нужно написать программу, которая будет на основе хранимых данных исполнять пользовательские команды.

Исходные данные имеют следующую структуру:

1. перечень всех документов
```
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
```
2. перечень полок, на которых хранятся документы (если документ есть в documents, то он обязательно должен быть и в directories)
```
directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}
```

Общие требования к программе:
- код должен быть грамотно декомпозирован (каждая функция отвечает за свою конкретную задачу, дублирующийся функционал переиспользуется, а его код не повторяется);
- в коде отсутствуют глобальные переменные (за исключением **documents** и **directories**);
- пользовательский ввод обрабатывается в цикле **while** до тех пор, пока пользователь явно не завершит программу (вводом команды "**q**").

````python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}

def p(number):
    name = ''
    for item in documents:
        if item['number'] == number:
            name = item['name']
    if name != '':
        return name
        print(f'Владелец документа: {name}')
    else:
        print('Документ не найден в базе')

def s(number):
    shelfnumber = 0
    for shelf in directories:
        if number in directories[shelf]:
            shelfnumber = shelf
    return shelfnumber
    
def fullinfo():
    for item in documents:
        print(f'№: {item["number"]}, тип: {item["type"]}, владелец: {item["name"]}, полка хранения: {s(item["number"])}')
        
# №: 2207 876234, тип: passport, владелец: Василий Гупкин, полка хранения: 1
# №: 11-2, тип: invoice, владелец: Геннадий Покемонов, полка хранения: 1
# №: 10006, тип: insurance, владелец: Аристарх Павлов, полка хранения: 2
        
while 1 > 0:
    command = input('Введите команду: ')
    if command == 'q':
        print('stop')
        break
    if command == 'p':
        number = input('Введите номер документа: ')
        print(p(number))
    if command == 's':
        number = input('Введите номер документа: ')
        shelfnumber = s(number)
        if shelfnumber == 0:
            print('Документ не найден в базе')
        else:
            print(f'Документ хранится на полке: {shelfnumber}')
    if command == '/':
        fullinfo()
````

````python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}

def p(number):
    name = ''
    for item in documents:
        if item['number'] == number:
            name = item['name']
    return name
    
def s(number):
    shelfnumber = 0
    for shelf in directories:
        if number in directories[shelf]:
            shelfnumber = shelf
    return shelfnumber
    
def fullinfo():
    for item in documents:
        print(f'№: {item["number"]}, тип: {item["type"]}, владелец: {item["name"]}, полка хранения: {s(item["number"])}')

def addshelf(shelfnumber):
    if shelfnumber in directories.keys():
        print(f'Такая полка уже существует. Текущий перечень полок: {", ".join(directories.keys())}')
    else:
        directories[shelfnumber] = []
        print(f'Полка добавлена. Текущий перечень полок: {", ".join(directories.keys())}')

def removeshelf(shelfnumber):
    if shelfnumber in directories.keys():
        if len(directories[shelfnumber]) > 0:
            print(f'На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: {", ".join(directories.keys())}')
        else:
            directories.pop(shelfnumber)
            print(f'Полка удаленa. Текущий перечень полок: {", ".join(directories.keys())}')
    else:
        print(f'Такой полки не существует. Текущий перечень полок: {", ".join(directories.keys())}')

        
while 1 > 0:
    command = input('Введите команду: ')
    if command == 'q':
        print('stop')
        break
    if command == 'p':
        number = input('Введите номер документа: ')
        name = p(number)
        if name != '':
            print(f'Владелец документа: {name}')
        else:
            print('Документ не найден в базе')
    if command == 's':
        number = input('Введите номер документа: ')
        shelfnumber = s(number)
        if shelfnumber == 0:
            print('Документ не найден в базе')
        else:
            print(f'Документ хранится на полке: {shelfnumber}')
    if command == '/':
        fullinfo()
    if command == 'ads':
        shelfnumber = input('Введите номер полки: ')
        addshelf(shelfnumber)
    if command == 'ds':
        shelfnumber = input('Введите номер полки: ')
        removeshelf(shelfnumber)
````
