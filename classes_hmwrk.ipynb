```python
class Rate:
    def __init__(self, format_='value'):
        self.format = format_
    
    def exchange_rates(self):
        """
        Возвращает ответ сервиса с информацией о валютах в виде:
        
        {
            'AMD': {
                'CharCode': 'AMD',
                'ID': 'R01060',
                'Name': 'Армянских драмов',
                'Nominal': 100,
                'NumCode': '051',
                'Previous': 14.103,
                'Value': 14.0879
                },
            ...
        }
        """
        self.r = requests.get('https://www.cbr-xml-daily.ru/daily_json.js')
        return self.r.json()['Valute']
    
    def make_format(self, currency):
        """
        Возвращает информацию о валюте currency в двух вариантах:
        - полная информация о валюте при self.format = 'full':
        Rate('full').make_format('EUR')
        {
            'CharCode': 'EUR',
            'ID': 'R01239',
            'Name': 'Евро',
            'Nominal': 1,
            'NumCode': '978',
            'Previous': 79.6765,
            'Value': 79.4966
        }
        
        Rate('value').make_format('EUR')
        79.4966
        """
        response = self.exchange_rates()
        
        if currency in response:
            if self.format == 'full':
                return response[currency]
            
            if self.format == 'value':
                return response[currency]['Value']
        
        return 'Error'
    
    def eur(self, diff=False):
        if diff == True:
            """Возвращает разницу: сегодняшний курс минус вчерашний"""
            return self.calculate_diff('EUR')
        """Возвращает курс евро на сегодня в формате self.format"""
        return self.make_format('EUR')
    
    def usd(self, diff=False):
        if diff == True:
            """Возвращает разницу: сегодняшний курс минус вчерашний"""
            return self.calculate_diff('USD')
        """Возвращает курс доллара на сегодня в формате self.format"""
        return self.make_format('USD')
    
    def brl(self, diff=False):
        if diff == True:
            """Возвращает разницу: сегодняшний курс минус вчерашний"""
            return self.calculate_diff('BRL')
        """Возвращает курс бразильского реала на сегодня в формате self.format"""
        return self.make_format('BRL')
    
    # Задание 1
    def return_name_of_max_rate(self):
        jsonvalutes = r.json()['Valute']
        max_rate = 0
        name_for_return = ''
        for valute in jsonvalutes:
            if int(jsonvalutes[valute]['Value']) > max_rate:
                max_rate = int(jsonvalutes[valute]['Value'])
                name_for_return = jsonvalutes[valute]['Name']
        return name_for_return
    
    # Задание 2
    def calculate_diff(self, valute):
        return float(r.json()['Valute'][valute]['Value']) - float(r.json()['Valute'][valute]['Previous'])
    
testrate = Rate()
print(testrate.return_name_of_max_rate())
testrate.usd(diff=True)
```
